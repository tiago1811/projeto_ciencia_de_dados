```python
# Importar bibliotecas necessárias
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import probplot, normaltest

# Carregar o dataset
df = pd.read_csv('escolas.mod.csv')

# Definir variáveis quantitativas
quantitative_vars = [
    "prop_aprovados_anos_inciais_ef", "prop_reprovados_anos_iniciais_ef", "prop_abandono_anos_iniciais_ef", 
    "prop_aprovados_anos_finais_ef", "prop_reprovados_anos_finais_ef", "prop_abandono_anos_finais_ef"
]

# Função para análise de distribuição (corrigida)
def analyze_distribution(data, column):
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))

    # Histograma e densidade
    sns.histplot(data[column], kde=True, ax=axes[0], color='skyblue')
    axes[0].set_title(f'Histograma e Densidade: {column}')
    axes[0].set_xlabel(column)
    axes[0].set_ylabel('Frequência')

    # Q-Q Plot
    probplot(data[column].dropna(), dist="norm", plot=axes[1])
    axes[1].set_title(f'Q-Q Plot: {column}')

    # Teste de Normalidade
    stat, p_value = normaltest(data[column].dropna())
    is_normal = "Sim" if p_value > 0.05 else "Não"

    # Tabela de valores e teste
    summary_stats = {
        "Média": data[column].mean(),
        "Mediana": data[column].median(),
        "Moda": data[column].mode().iloc[0] if not data[column].mode().empty else None,
        "Mínimo": data[column].min(),
        "Máximo": data[column].max(),
        "Variância": data[column].var(),
        "Amplitude": data[column].max() - data[column].min(),
        "Teste de Normalidade (p > 0.05)": is_normal,
    }

    # Ajustando o formato de cellText para lista de listas
    cell_text = [[value] for value in summary_stats.values()]

    # Exibir resultados em tabela
    axes[2].axis('off')
    axes[2].table(cellText=cell_text, 
                  rowLabels=list(summary_stats.keys()), 
                  colLabels=["Estatísticas"], 
                  loc='center', fontsize=12)
    axes[2].set_title(f'Estatísticas e Teste de Normalidade: {column}')

    plt.tight_layout()
    plt.show()

# Analisar cada variável quantitativa
for col in quantitative_vars:
    analyze_distribution(df, col)
```
